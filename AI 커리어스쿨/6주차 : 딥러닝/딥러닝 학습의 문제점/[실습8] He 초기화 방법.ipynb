```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)
    
def relu(x):
    result = np.maximum(0,x)
    return result

'''
1.   .

2.        .
   Numpy     .
   
3. relu   'a_relu' .
'''

def main():
    
    x_relu = np.random.randn(1000,100)
    
    node_num = 100
    hidden_layer_size = 5
    
    activations_relu = {}
    
    for i in range(hidden_layer_size):
        if i != 0:
            x_relu = activations_relu[i-1]
            
        w_relu = np.random.randn(100,100) * (np.sqrt(2)/np.sqrt(node_num))
        
        a_relu = np.dot(x_relu , w_relu)
        
        z_relu = relu(a_relu)
        
        activations_relu[i] = z_relu
    
    return activations_relu    


def visualize_activations(activations, title):
    plt.figure(figsize=(15, 10))
    for i, layer_output in activations.items():
        plt.subplot(1, len(activations), i + 1)
        plt.hist(layer_output.flatten(), bins=50, range=(0, 1))
        plt.title(f"Layer {i + 1}")
        plt.xlabel("Activation")
        plt.ylabel("Frequency")
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    activations_1 = main()
    visualize_activations(activations_1, "Activation Distributions")
```
