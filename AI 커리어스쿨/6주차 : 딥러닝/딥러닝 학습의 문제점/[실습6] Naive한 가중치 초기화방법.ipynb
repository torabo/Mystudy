```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(100)

def sigmoid(x):
    result = 1 / (1 + np.exp(-x))
    return result

'''
1. 입력 데이터를 정의하세요.

2. 가중치를 정의하세요.

3. sigmoid를 통과할 값인 'a_1', 'a_2'를 정의하세요.
'''

def main():
    
    x_1 = np.random.randn(1000,100)
    x_2 = np.random.randn(1000,100)
    
    node_num = 100
    hidden_layer_size = 5
    
    activations_1 = {}
    activations_2 = {}
    
    for i in range(hidden_layer_size):
        if i != 0:
            x_1 = activations_1[i-1]
            x_2 = activations_2[i-1]
        
        w_1 = np.random.randn(100,100)
        w_2 = np.random.randn(100,100)*0.01 + 0
        
        a_1 = np.dot(x_1 , w_1)
        a_2 = np.dot(x_2 , w_2)
        
        z_1 = sigmoid(a_1)
        z_2 = sigmoid(a_2)
        
        activations_1[i] = z_1
        activations_2[i] = z_2

    
    return activations_1, activations_2


def visualize_activations(activations, title):
    plt.figure(figsize=(15, 10))
    for i, layer_output in activations.items():
        plt.subplot(1, len(activations), i + 1)
        plt.hist(layer_output.flatten(), bins=50, range=(0, 1))
        plt.title(f"Layer {i + 1}")
        plt.xlabel("Activation")
        plt.ylabel("Frequency")
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    activations_1, activations_2 = main()
    visualize_activations(activations_1, "Activation Distributions for x_1")
    visualize_activations(activations_2, "Activation Distributions for x_2")
```
