# [실습] Matplotlib 기본 문법

## 1. 실습 준비하기

먼저 실습에 필요한 패키지 불러옵니다.
matplotlib.pyplot는 데이터 시각화를 위한 다양한 함수들을 제공합니다.

아래 코드에서는 해당 라이브러리를 `plt`라는 이름으로 import 합니다.

```python
import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="AppleGothic")
```

다음 간단한 그래프를 그려보기 위한 데이터를 준비합니다.

이번 실습에서는 각 연도별로 프로그래밍 언어의 점유율을 나타낸 가상의 데이터를 사용합니다.

```python
# 데이터 리스트로 생성
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
```

## 2. 그래프 그리기

간단한 그래프를 그릴 때는 `plt`의 함수들을 활용하면 간단하게 그릴 수 있습니다.

가장 간단하게 연도별 파이썬의 점유율을 선 그래프로 표현해 보겠습니다.

`plt.plot()` 함수는 선 그래프를 그리는데 사용하는 함수입니다.

가장 간단하게 선 그래프를 그리기 위해서는 `plot()` 함수에 x, y 데이터를 순서대로 전달하면 됩니다.

선 그래프의 자세한 사용법은 뒤의 실습에서 다시 다루겠습니다.

```python
plt.plot(year, python)  # x축은 년도, y축은 파이썬의 점유율
plt.show() # 만든 그래프를 화면에 보여주는 함수.(주피터 환경에서는 생략해도 그래프가 나타납니다.)
```

간단하게 연도별 파이썬의 점유율 그래프를 그렸습니다.

이 그래프가 어떤 그래프인지 표현하기 위해서 제목을 붙여줍니다.

그래프의 제목은 `title()` 함수를 이용하여 설정할 수 있습니다.

```python
plt.title("파이썬의 언어 점유율")  # 그래프의 제목을 설정
plt.plot(year, python)  # x축은 년도, y은 파이썬의 점유율
```

**축의 레이블 표시하기**

이번에는 각 축이 무엇을 의미하는지 알려주기 위해 각 축의 레이블을 표시하겠습니다.

x축과 y축의 레이블을 설정하려면 각각 `xlabel()` 함수와 `ylabel()` 함수를 사용해야 합니다.

```python
plt.title("파이썬의 언어 점유율")
plt.xlabel("연도")  # x축의 레이블을 "연도"로 설정
plt.ylabel("점유율")  # y축의 레이블을 "점유율"로 설정
plt.plot(year, python)
```

**텍스트의 크기 변경하기**

필요에 따라 텍스트의 크기를 변경할 수 있습니다.

`fontsize` 매개변수를 전달하면 텍스트의 크기를 변경할 수 있습니다.

제목과 레이블의 크기를 변경해 보겠습니다.

```python
plt.title("파이썬의 언어 점유율", fontsize=20)  # 제목의 크기를 20으로 설정
plt.xlabel("연도", fontsize=15)  # 레이블의 크기를 15로 설정
plt.ylabel("점유율", fontsize=15)  # 레이블의 크기를 15로 설정
plt.plot(year, python)
```

**여러 그래프를 함께 표시하기**

다른 언어의 점유율과 함께 비교하기 위해서 그래프를 추가해 보겠습니다.

`plot()` 함수는 호출할 때마다 하나의 선 그래프를 그려줍니다.

다른 선 그래프를 추가하면 여러 번 `plot()`을 호출하면 됩니다.

```python
plt.title("프로그래밍 언어 점유율", fontsize=20)
plt.xlabel("연도", fontsize=15)
plt.ylabel("점유율", fontsize=15)
plt.plot(year, python)

# 다른 언어의 그래프를 추가
plt.plot(year, C)
plt.plot(year, java)
plt.plot(year, JS)
```

**그래프의 범례 (Legend) 표시하기**

여러 선 그래프를 그리니 각 선이 어떤 언어의 그래프인지 구분하기 어려워졌습니다.

이렇게 하나의 그래프에 여러 데이터를 표시할 경우 범례 (Legend)를 추가할 필요가 있습니다.

범례 (Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트입니다.

`legend()`함수를 사용해서 그래프에 범례를 표시할 수 있습니다.

그래프의 이름을 표시하는 첫 번째 방법은 `legend()`함수에 각 그래프의 이름을 순서대로 전달하는 것입니다.

```python
plt.title("프로그래밍 언어 점유율", fontsize=20)
plt.xlabel("연도", fontsize=15)
plt.ylabel("점유율", fontsize=15)

plt.plot(year, python)
plt.plot(year, C)
plt.plot(year, java)
plt.plot(year, JS)

plt.legend(['Python', 'C', 'JAVA', 'Java Script']) # 그래프를 그린 순서대로 이름을 전달
```

위 방법으로 문제없이 그래프의 이름을 출력했지만, `plot()`을 호출하는 순서에 영향을 받기에 좋은 방법은 아닙니다.

이런 경우 `plot()`을 통해 그래프를 그릴 때 각 그래프의 `label` 문자열을 지정하여 설정할 수 있습니다.

```python
plt.title("프로그래밍 언어 점유율", fontsize=20)
plt.xlabel("연도", fontsize=15)
plt.ylabel("점유율", fontsize=15)

# 각 그래프의 label을 설정
plt.plot(year, python, label="Python")
plt.plot(year, C, label="C")
plt.plot(year, java, label="JAVA")
plt.plot(year, JS, label="Java Script")

plt.legend()  # 각 그래프의 label이 이미 전달되었으니 따로 전달하지 않고 호출합니다.
```

## 3. figure

지금까지의 그래프는 모두 `plt`의 함수들을 직접 호출하면서 그렸습니다.

이는 간단한 그래프를 그리며 하나의 그래프만 사용하기 때문에 가능했습니다.

실제로는 여러 개의 그래프를 관리해야 하는 경우가 많이 있는데, 그럴 경우 `subplot`을 이용해야 합니다.

subplot에 대해 알아보기 전에 그래프의 도화지에 해당하는 figure에 대해서 알아보겠습니다.

먼저 아래의 코드를 실행해보겠습니다.

```python
fig1 = plt.figure(1,figsize=(4,3))  # ID가 1인 새로운 figure 객체를 생성하고 선택합니다. 앞으로 호출하는 함수는 fig1에 적용됩니다.
plt.plot(year, python, label="Python")  # 현재 선택된 figure에 선 그래프를 그립니다.
plt.title("fig1")  # 현재 선택된 figure의 제목을 설정합니다.

fig2 = plt.figure(2,figsize=(4,3))  # ID가 2인 새로운 figure 객체를 생성하고 선택합니다. 앞으로 호출하는 함수는 fig2에 적용됩니다.
plt.plot(year, java, label="JAVA")  # 현재 선택된 figure에 선 그래프를 그립니다.
plt.title("fig2")  # 현재 선택된 figure의 제목을 설정합니다.

plt.show()  # 모든 figure를 화면에 그립니다.
```

그래프가 두개 나타났습니다. matplotlib의 `figure`는 그래프를 그릴 종이에 해당하는 객체입니다. 위 코드는 두개의 figure에 각각 파이썬과 자바의 그래프를 그렸습니다.

`plt.figure`를 통해 figure를 생성하고 선택했습니다.

각각에 1, 2와 같은 ID를 부여했으며 `fizsize` 매개 변수를 전달하여 크기가 3, 4인치가 되도록 설정했습니다.

마지막 줄의 `plt.show()`를 호출하면 모든 figure를 출력하게 됩니다.

각각의 그래프는 `fig1`, `fig2`를 통해 확인할 수 있습니다.

```python
fig2 # fig2의 그래프만 보기
```

또한, 우리는 figure을 생성하면서 1, 2 와 같은 ID를 부여했기 때문에 해당 값을 통해 다시 활성화 시킬 수 있습니다.

아래 코드는 figure를 바꿔가면서 그래프를 그립니다.

```python
fig1 = plt.figure(1, figsize=(4, 3))  # ID가 1인 새로운 figure 객체를 생성하고 선택합니다. 앞으로 호출하는 함수는 fig1에 적용됩니다.
plt.plot(year, python, label="Python")  # 현재 선택된 figure에 선 그래프를 그립니다.
plt.title("fig1")  # 현재 선택된 figure의 제목을 설정합니다.

fig2 = plt.figure( 2, figsize=(4, 3))  # ID가 2인 새로운 figure 객체를 생성하고 선택합니다. 앞으로 호출하는 함수는 fig2에 적용됩니다.
plt.title("fig2")  # 현재 선택된 figure의 제목을 설정합니다.

plt.figure(1)  # ID가 1인 figure를 선택합니다.(없으면 새로 만듭니다.)
plt.plot(year, java, label="JAVA")  # 현재 선택된 figure에 선 그래프를 그립니다.

plt.show()
```

결과를 보면 `fig1`에는 2개의 선 그래프가 그려져 있습니다.

먼저 `fig1`에 그래프를 그리고, 그 다음에 `fig2`의 작업을 수행한 후에 다시 `fig1`로 돌아와서 추가적인 자바 그래프를 그렸습니다.

이렇게 figure를 여러 개 생성하면 종이를 여러 장 사용하는 것과 같이 여러 개의 그래프를 그릴 수 있습니다.

## 4. subplot

실제로 그래프를 그릴 때, 두 개의 그래프를 각각 다른 종이에 그릴 수도 있지만, 때로는 두 그래프를 함께 비교해야 할 때가 있어서 하나의 종이에 나란히 그리기도 합니다.

이런 경우에는 하나의 `figure`에서 여러 개의 그래프를 그리기 위해 `subplots`을 사용합니다.

```python
fig, (ax1, ax2) = plt.subplots(1, 2) # 세로로 1칸, 가로로 2칸 총 2개의 subplot을 생성합니다.
# 생성된 subplot은 각각 ax1과 ax2에 저장됩니다.

ax1.plot(year, python)  # ax1에 파이썬 선 그래프를 그립니다.
ax2.plot(year, java)  # ax1에 자바 선 그래프를 그립니다.

plt.show() # 그래프를 표시합니다.
```

위의 코드에서는 `plt.subplots(1, 2)`를 사용하여 하나의 figure에 두 개의 subplot을 생성합니다.

`1`, `2`는 각각 행과 열의 개수를 나타냅니다. 따라서 이 경우에는 1행 2열의 subplot이 생성됩니다.

이렇게 `subplot`을 생성하면 하나의 종이에 여러 개의 그래프를 그릴 수 있습니다.

반대로 `plt.subplots(2, 1)`를 통 2행 1열로 그래프를 배치할 수도 있습니다.

```python
fig, (ax1, ax2) = plt.subplots(2, 1)  # 2행 1열로 총 2개의 그래프를 생성.
# 생성된 subplot은 각각 ax1과 ax2에 저장됩니다.

ax1.plot(year, python)  # 현재 선택된 figure에 선 그래프를 그립니다.
ax2.plot(year, java)  # 현재 선택된 figure에 선 그래프를 그립니다.

plt.show()  # 그래프를 표시합니다.
```

이렇게 생성된 subplot `ax1`, `ax2`은 서브플롯(subplot)이라고 부르며, 위에서 `plt`에서 사용하던 대부분의 함수들을 비슷한 형식으로 사용할 수 있습니다.

예를 들어 위에서는 `plt.title()`함수를 통해 서브플롯의 제목을 설정했다면 이제는 `ax1.set_title()`함수를 이용해 제목을 설정해야 합니다.

```python
ax1.set_title("ax1의 제목") # ax1의 제목을 설정
ax2.set_title("ax2의 제목") # ax2의 제목을 설정
fig
```

하지만 이렇게 출력하니 두 서브플롯의 일부가 겹치게 생성되었습니다.

각 서브플롯의 간격을 조절하는 방법은 다양한 방법이 있지만, 자동으로 조절하는 방법을 지원합니다.

처음에 `subplot()`함수를 이용해 서브플롯을 생성할 때 `constrained_layout=True`를 매개변수에 전달하면 서브플롯 간의 간격을 최적의 수치로 자동으로 조절해 줍니다.

```python
fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True)  # constrained_layout=True를 전달
ax1.plot(year, python)  # ax1에 파이썬 선 그래프를 그립니다.
ax1.set_title("ax1의 제목") # ax1의 제목을 설정
ax1.set_xlabel("x라벨") # ax1의 xlabel을 설정

ax2.plot(year, java)  # ax2에 자바 선 그래프를 그립니다.
ax2.set_title("ax2의 제목") # ax2의 제목을 설정

plt.show()  # 그래프를 표시합니다.
```
