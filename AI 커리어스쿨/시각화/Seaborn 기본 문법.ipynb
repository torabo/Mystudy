# [실습] Seaborn 기본 문법

Seaborn은 Matplotlib 위에 구축된 Python 라이브러리입니다. Matplotlib보다 간단한 코드로 그래프를 그릴 수 있는 함수들을 제공합니다.

Seaborn을 이용해 그래프를 그리는 것이 더 편하고 간결하지만, 일부 그래프를 그리는 경우에는 Matplotlib를 사용해야 할 수 있습니다.

## 1. 실습 준비하기

이번 실습에서도 실습에 필요한 패키지 불러옵니다.

아래 코드는 `seaborn` 라이브러리를 `sns`라는 이름으로 import 합니다.

```python
import seaborn as sns
```

여기에 각종 설정을 위해 matplotlib도 불러옵니다.

matplotlib위에서 개발된 라이브러리인 만큼 여러가지 옵션을 조정하려면 모듈을 불러올 필요가 있습니다.

아래 코드는 `matplotlib`를 불러오고 한글의 출력을 위해 폰트를 변경해주는 코드입니다.

```python
import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="AppleGothic")
```

이번 실습에서도 전 실습과 같은 예시 데이터를 사용하겠습니다.

이번 실습에서는 각 연도별로 프로그래밍 언어의 점유율을 나타낸 가상의 데이터를 사용합니다.

```python
# 데이터 리스트로 생성
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
```

## 2. 그래프 그리기 

seaborn에서 그래프를 그릴 때에도 `sns`의 함수들을 호출해 간단한 그래프를 그릴 수 있습니다.

가장 간단하게 연도별 파이썬 점유율을 선 그래프로 표현하겠습니다.

`sns.lineplot()` 함수는 선 그래프를 그리는데 사용하는 함수입니다.

`x`와 `y` 매개변수에 데이터를 전달하면 간단하게 그래프를 그릴 수 있습니다.

```python
sns.lineplot(x=year, y=python)  # 파이썬 그래프를 그리기
```

seaborn은 matplotlib를 기반으로 개발된 라이브러리이기 때문에 matplotlib의 매개변수 대부분을 사용할 수 있습니다.

아래 코드는 자바와 파이썬 그래프를 그리고 각각 그래프의 범례를 그리는 코드입니다.

```python
sns.lineplot(x=year, y=python, label="python")  # "python"이라는 레이블 추가
sns.lineplot(x=year, y=java, label="java")  # "java"라는 레이블 추가
```

`sns.lineplot()` 함수는 `matplotlib.axes` 를 반환하는 함수입니다.

이 말은 `matplotlib.axes`의 매소드들을 이용해 그래프의 모양을 조절할 수 있다는 말입니다.

아래 코드는 그래프를 그리면서 반환된 객체를 이용해 x축과 y축 레이블을 설정합니다.

```python
lp= sns.lineplot(x=year, y=python, label="python")

lp.set_xlabel("연도") # x축의 레이블을 설정 (matplotlib의 ax와 같은 방식)
lp.set_ylabel("점유율") # y축의 레이블을 설정 
```

## 3. 테마 사용하기

matplotlib의 기본 설정을 이용하면 하얀색 배경에 다소 투박한 그래프가 그려집니다.

seaborn에서는 몇가지 테마를 제공하는데, 이 테마를 사용하면 그래프를 원하는 스타일로 그릴 수 있습니다.

태마는 `sns.set_style("테마이름")`을 통해 지정할 수 있습니다.

```python
sns.set_style("darkgrid")  # 스타일을 darkgrid로 변경
lp = sns.lineplot(x=year, y=python, label="python")  # 그래프 그리기
```

스타일을 설정하면 다양한 유형으로 그래프의 디자인을 쉽게 변경할 수 있습니다.

## 4. subplot

seaborn은 matplotlib를 기반으로 개발되었기 때문에 subplot 역시 사용할 수 있습니다.

seaborn에서 그래프를 그리는 함수에 subplot으로 생성된 `ax`라는 매개변수로 전달하면 해당 서브플롯에 seaborn을 이용해 그래프를 그릴 수 있습니다.

아래 코드는 서브플롯을 생성하고 그 위에 seaborn을 이용해 그래프를 그리는 방법입니다.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)  # 1행 2열의 서브플롯을 생성합니다.

fig = sns.lineplot(x=year, y=python, label="python", ax=ax1)  # seaborn을 이용해 그래프 그리기
sns.lineplot(x=year, y=java, label="java", ax=ax2)  # seaborn을 이용해 그래프 그리기
```
